```python
# we need to find out optimal beta
def least_squares_fit(xs: List[la.Vector],
                      ys: List[float],
                      learning_rate: float = 0.001,
                      num_steps: int = 1000,
                      batch_size: int = 1) -> la.Vector:
    """
    Find the beta that minimizes the sum of squared errors
    assuming the model y = dot(x, beta).
    """
    # Start with a random guess
    guess = [random.random() for _ in xs[0]]

    for _ in range(num_steps):
        for start in range(0, len(xs), batch_size):
            batch_xs = xs[start:start+batch_size]
            batch_ys = ys[start:start+batch_size]

            gradient = la.vector_mean([sqerror_gradient(x, y, guess)
                                    for x, y in zip(batch_xs, batch_ys)])
            guess = gd.gradient_step(guess, gradient, -learning_rate)
    return guess

random.seed(0)
# I used trial and error to choose num_iters and step_size.
# This will run for a while.
learning_rate = 0.001

# every variable contribute to the y we want to predict, but the model does not capture the interaction between the variables.
beta = least_squares_fit(inputs, st.daily_minutes_good, learning_rate, 5000, 25)
assert 30.50 < beta[0] < 30.70  # constant
assert  0.96 < beta[1] <  1.00  # num friends
assert -1.89 < beta[2] < -1.85  # work hours per day
assert  0.91 < beta[3] <  0.94  # has PhD

def multiple_r_squared(xs: List[la.Vector], ys: la.Vector, beta: la.Vector) -> float:
    sum_of_squared_errors = sum(error(x, y, beta) ** 2
                                for x, y in zip(xs, ys))
    return 1.0 - sum_of_squared_errors / slr.total_sum_of_squares(ys)

# the more variables we add to the model the more the r squared will increase
assert 0.67 < multiple_r_squared(inputs, st.daily_minutes_good, beta) < 0.68
```

```python
from typing import TypeVar, Callable
# BOOTSTRAP 
# when you have data, you could want to know the median or custom function.
# how confident are you about the estimation?
# if you get 2 dataset: one with a lot of data close to 100 and one with data close to 0 and 200. you will have a simillar result
# to improve your estimate, you can bootstrap the calculation. you can repeatedly get new samples by picking replaces from the original data.
# recalculate the median everytime and at the end check the standard deviation of the collected medians.
X = TypeVar('X')        # Generic type for data
Stat = TypeVar('Stat')  # Generic type for "statistic"

def bootstrap_sample(data: List[X]) -> List[X]:
    """randomly samples len(data) elements with replacement"""
    return [random.choice(data) for _ in data]

def bootstrap_statistic(data: List[X],
                        stats_fn: Callable[[List[X]], Stat],
                        num_samples: int) -> List[Stat]:
    """evaluates stats_fn on num_samples bootstrap samples from data"""
    return [stats_fn(bootstrap_sample(data)) for _ in range(num_samples)]

# 101 points all very close to 100
close_to_100 = [99.5 + random.random() for _ in range(101)]

# 101 points, 50 of them near 0, 50 of them near 200
far_from_100 = ([99.5 + random.random()] +
                [random.random() for _ in range(50)] +
                [200 + random.random() for _ in range(50)])

medians_close = bootstrap_statistic(close_to_100, st.median, 100)
medians_far = bootstrap_statistic(far_from_100, st.median, 100)

# the standard deviations will be completly different
assert st.standard_deviation(medians_close) < 1
assert st.standard_deviation(medians_far) > 90
```

```python
def estimate_sample_beta(pairs: List[Tuple[la.Vector, float]]):
    x_sample = [x for x, _ in pairs]
    y_sample = [y for _, y in pairs]
    beta = least_squares_fit(x_sample, y_sample, learning_rate, 5000, 25)
    #print("bootstrap sample", beta)
    return beta

random.seed(0) # so that you get the same results as me

# This will take a couple of minutes!
bootstrap_betas = bootstrap_statistic(list(zip(inputs, st.daily_minutes_good)),
                                      estimate_sample_beta,
                                      100)

bootstrap_standard_errors = [
    st.standard_deviation([beta[i] for beta in bootstrap_betas])
    for i in range(4)]

print(bootstrap_standard_errors)

# [1.272,    # constant term, actual error = 1.19
#  0.103,    # num_friends,   actual error = 0.080
#  0.155,    # work_hours,    actual error = 0.127
#  1.249]    # phd,           actual error = 0.998

def p_value(beta_hat_j: float, sigma_hat_j: float) -> float:
    if beta_hat_j > 0:
        # if the coefficient is positive, we need to compute twice the
        # probability of seeing an even *larger* value
        return 2 * (1 - pr.normal_cdf(beta_hat_j / sigma_hat_j))
    else:
        # otherwise twice the probability of seeing a *smaller* value
        return 2 * pr.normal_cdf(beta_hat_j / sigma_hat_j)

assert p_value(30.58, 1.27)   < 0.001  # constant term
assert p_value(0.972, 0.103)  < 0.001  # num_friends
assert p_value(-1.865, 0.155) < 0.001  # work_hours
assert p_value(0.923, 1.249)  > 0.4    # phd
# the PHD coeffient is more rnadom than meaningful
```

```python
# REGULARIZATION
# we add a penality to the error term that gets bigger as beta gets larger.

# alpha is a *hyperparameter* controlling how harsh the penalty is.
# Sometimes it's called "lambda" but that already means something in Python.
def ridge_penalty(beta: la.Vector, alpha: float) -> float:
    return alpha * la.dot(beta[1:], beta[1:])

def squared_error_ridge(x: la.Vector,
                        y: float,
                        beta: la.Vector,
                        alpha: float) -> float:
    """estimate error plus ridge penalty on beta"""
    return error(x, y, beta) ** 2 + ridge_penalty(beta, alpha)

def ridge_penalty_gradient(beta: la.Vector, alpha: float) -> la.Vector:
    """gradient of just the ridge penalty"""
    return [0.] + [2 * alpha * beta_j for beta_j in beta[1:]]

def sqerror_ridge_gradient(x: la.Vector,
                           y: float,
                           beta: la.Vector,
                           alpha: float) -> la.Vector:
    """
    the gradient corresponding to the ith squared error term
    including the ridge penalty
    """
    return la.add(sqerror_gradient(x, y, beta),
               ridge_penalty_gradient(beta, alpha))

# same as leas_squares_fit but using sqerror_ridge_gradient
def least_squares_fit_ridge(xs: List[la.Vector],
                      ys: List[float],
                      alpha: float,
                      learning_rate: float = 0.001,
                      num_steps: int = 1000,
                      batch_size: int = 1) -> la.Vector:
    """
    Find the beta that minimizes the sum of squared errors
    assuming the model y = dot(x, beta).
    """
    # Start with a random guess
    guess = [random.random() for _ in xs[0]]

    for _ in range(num_steps):
        for start in range(0, len(xs), batch_size):
            batch_xs = xs[start:start+batch_size]
            batch_ys = ys[start:start+batch_size]

            gradient = la.vector_mean([sqerror_ridge_gradient(x, y, guess, alpha)
                                    for x, y in zip(batch_xs, batch_ys)])
            guess = gd.gradient_step(guess, gradient, -learning_rate)
    return guess

random.seed(0)
beta_0 = least_squares_fit_ridge(inputs, st.daily_minutes_good, 0.0,  # alpha
                                 learning_rate, 5000, 25)
# [30.51, 0.97, -1.85, 0.91]
assert 5 < la.dot(beta_0[1:], beta_0[1:]) < 6
assert 0.67 < multiple_r_squared(inputs, st.daily_minutes_good, beta_0) < 0.69

# as we increase alpha, the goodness of fit gets worse, but the size of beta gets smaller:
beta_0_1 = least_squares_fit_ridge(inputs, st.daily_minutes_good, 0.1,  # alpha
                                   learning_rate, 5000, 25)
# [30.8, 0.95, -1.83, 0.54]
assert 4 < la.dot(beta_0_1[1:], beta_0_1[1:]) < 5
assert 0.67 < multiple_r_squared(inputs, st.daily_minutes_good, beta_0_1) < 0.69

beta_1 = least_squares_fit_ridge(inputs, st.daily_minutes_good, 1,  # alpha
                                 learning_rate, 5000, 25)
# [30.6, 0.90, -1.68, 0.10]
assert 3 < la.dot(beta_1[1:], beta_1[1:]) < 4
assert 0.67 < multiple_r_squared(inputs, st.daily_minutes_good, beta_1) < 0.69

beta_10 = least_squares_fit_ridge(inputs, st.daily_minutes_good,10,  # alpha
                                  learning_rate, 5000, 25)
# [28.3, 0.67, -0.90, -0.01]
assert 1 < la.dot(beta_10[1:], beta_10[1:]) < 2
assert 0.5 < multiple_r_squared(inputs, st.daily_minutes_good, beta_10) < 0.6
# we have the confirm that the PHD coefficient is not meaningful. PHD coefficient vanishes as we increase the penality.

#another approach
def lasso_penalty(beta, alpha):
    return alpha * sum(abs(beta_i) for beta_i in beta[1:])
```
